## 2. Measure of central tendency (Mean, Median & Mode)

```python
# import the  Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the dataset
loc = "C:\\Users\\amira\\Data Science\\0. Some  Basic Work\\07. Preprocessing\\Data\\train.csv"
dataset = pd.read_csv(loc)
```

```python
# Display all rows and col
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# see the first three rows
dataset.head(3)
```

```python
# how many rows and col
dataset.shape
```

```python
# dataset information
dataset.info()
```

```python
# check the how many missing value in each individual col
dataset.isnull().sum()
```

```python
### Check the each missing value of col in percentage
missing_value_per = dataset.isnull().sum() / dataset.shape[0] * 100
missing_value_per
```

```python
# take the col who have more than 20% missing values
missing_value_clm_gre_20 = missing_value_per[missing_value_per > 20].keys()
missing_value_clm_gre_20
```

```python
# drop the col who has more than 20% missing values
df2_drop_clm = dataset.drop(columns=missing_value_clm_gre_20)
df2_drop_clm.shape        # 81 - 6 = 75 col
```

```python
# take only numerical col including integer and float
df3_num=df2_drop_clm.select_dtypes(include=['int64','float64'])
df3_num.head()
```

```python
# visualize numerical missing on seaborn
plt.figure(figsize = (16, 9))
sns.heatmap(df3_num.isnull())
```

```python
# check missing missing
df3_num.isnull().sum()
```

```python
# take only missing value col
missing_num_var = [var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_var
```

```python
# see the data distribution before apply mean, median
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':5, 'color':'#DC143C'})
```

```python
# mean
df4_num_mean = df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
# compare orginal with mean using distplot
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df4_num_mean[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    plt.legend()
```

```python
# median
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
# orginal vs mean vs median
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20,hist=False, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df4_num_mean[var], bins=20,hist=False, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    sns.distplot(df5_num_median[var], bins=20,hist=False, kde_kws={'linewidth':3, 'color':'k'},label="Median",)
    plt.legend()
```

```python
# boxplot of orginal, mean, median
for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(dataset[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_num_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df5_num_median[var])
```

```python
# concatinate orginal, mean, median
df_concat = pd.concat([df3_num[missing_num_var],df4_num_mean[missing_num_var],df5_num_median[missing_num_var]], axis=1)
```
