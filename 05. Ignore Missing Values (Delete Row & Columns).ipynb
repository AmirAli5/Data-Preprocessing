## 1. Ignore Missing Values Row ( Delete Row & Columns)

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import House Prices dataset
loc = "C:\\Users\\amira\\Data Science\\0. Some  Basic Work\\07. Preprocessing\\Data\\train.csv"
dataset = pd.read_csv(loc)
```

```python
# Display all the rows and col of our dataset
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Check how many missing value in each individual col
dataset.isnull().sum()
```

```python
### Check the missing value in heatmap using seaborn
plt.figure(figsize = (25, 25))
sns.heatmap(dataset.isnull())
```

```python
### Check the each missing value of col in percentage
null_var = dataset.isnull().sum()/dataset.shape[0] * 100
null_var
```

```python
# drop the col which have more than 17% missing values
drop_columns = null_var[null_var > 17].keys()
drop_columns   # so 6 col is drop which have more than 17 missing values
```

```python
dataset_drop_col = dataset.drop(columns = drop_columns)
dataset_drop_col.shape            # 81 - 6 = 75
```

```python
### check again by visualizing
plt.figure(figsize = (25, 25))
sns.heatmap(dataset_drop_col.isnull())
```

```python
# Now drop the missing values
dataset_drop_row = dataset_drop_col.dropna()
dataset_drop_row.shape
```

```python
### check again by visualizing after drop rows (it's clear no more missing values)
plt.figure(figsize = (25, 25))
sns.heatmap(dataset_drop_row.isnull())data
```

```python
# check how many missing values
dataset_drop_row.isnull().sum().sum()
```

```python
# now take only categorical(obj) 
dataset_drop_row.select_dtypes(include = ['object']).columns
```

```python
pd.concat([dataset['MSZoning'].value_counts()/dataset.shape[0] * 100, 
          dataset_drop_row['MSZoning'].value_counts()/dataset_drop_row.shape[0] * 100], axis=1,
         keys=['MSZoning_org', 'MSZoning_clean'])
```

```python
def cat_var_dist(var):
    return pd.concat([dataset[var].value_counts()/dataset.shape[0] * 100, 
          dataset_drop_row[var].value_counts()/dataset_drop_row.shape[0] * 100], axis=1,
         keys=[var+'_org', var+'_clean'])
```
