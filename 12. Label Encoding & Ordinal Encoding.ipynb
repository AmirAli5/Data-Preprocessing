## Label Encoding & Ordinal Encoding

```python
# Import Libraries
import pandas as pd 
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
# Import Dataset
loc = "C:\\Users\\amira\\Data Science\\0. Some  Basic Work\\07. Preprocessing\\Data\\train.csv"
df = pd.read_csv(loc)
df.head(3)
```

```python
# Display all Row and Col
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# take only two variablesfor LE and OE 
df2 = df[["KitchenQual","BldgType"]]
```

```python
df2.head()
```

### Label Encoding

```python
# Initialize the Label Enoder
le = LabelEncoder()
```

```python
df2["BldgType_L_enc"] = le.fit_transform(df2["BldgType"])
```

```python
df2[["BldgType_L_enc","BldgType"]]
```

### Ordinal Encoding

```python
df["KitchenQual"].value_counts()
```

```python
"""# KitchenQual: Kitchen quality

       Ex	Excellent
       Gd	Good
       TA	Typical/Average
       Fa	Fair
"""
```

```python
order_Label = {"Ex":4,"Gd":3,"TA":2,"Fa":1}
```

```python
df2["KitchenQual_org_enc"] = df2["KitchenQual"].map(order_Label)  # here we create new
```
