```python
# same as but whole class
df_copy_mean = df.copy()
for var_class in df['LotConfig'].unique():
    df_copy_mean.update(df[df.loc[:,'LotConfig'] == var_class]["LotFrontage"].replace(np.nan,df[df.loc[:,'LotConfig'] == var_class]["LotFrontage"].mean()))
```

```python
# No missing value in LotFrontage col
df_copy_mean.isnull().sum()
```

```python
# Now take all the col with correspond col to compute and fill missing value using mean method
df_copy_mean = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy_mean.update(df[df.loc[:,cat_var] == var_class][num_var_miss].replace(np.nan,df[df.loc[:,cat_var] == var_class][num_var_miss].mean()))
```

```python
# check the col that we fill through
df_copy_mean[num_vars_miss].isnull().sum()
```

```python
# same step but using median method
df_copy_median = df.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in df[cat_var].unique():
        df_copy_median.update(df[df.loc[:,cat_var] == var_class][num_var_miss].replace(np.nan,df[df.loc[:,cat_var] == var_class][num_var_miss].median()))
```

```python
df_copy_median[num_vars_miss].isnull().sum()
```
